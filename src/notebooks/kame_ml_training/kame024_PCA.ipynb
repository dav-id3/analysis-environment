```python
import seaborn as sns
df = sns.load_dataset("iris")
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

X = df.loc[:, df.columns != "species"]
y = df["species"].values
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
X
```

```python
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
pca = PCA(n_components=2)
pca.fit(X)
X_pc = pca.transform(X)

y=y.reshape(-1, 1)
df_pc = pd.DataFrame(np.concatenate([X_pc, y], axis=1), columns=["first component", "second component", "species"]).astype({"first component": float, "second component": float})
```

```python
sns.scatterplot(x=df_pc["first component"], y=df_pc["second component"], hue=df_pc["species"])
```

```python
pca = PCA(n_components=4)
pca.fit(X)
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```
